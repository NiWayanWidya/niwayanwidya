```python
#KORELASI PEARSON

import pandas as pd
from scipy.stats import pearsonr

# Baca file Excel dari sheet 3
data = pd.read_excel('/content/Data Korelasi Pearson [www.spssindonesia.com].xlsx', sheet_name='Sheet3')

# Buat DataFrame
df = pd.DataFrame(data)

# Mengubah nama kolom menjadi huruf kecil agar sesuai dengan data
df.columns = df.columns.str.lower()

# Hitung korelasi Pearson antara X1 dan Y
correlation_x1_y, p_value_x1_y = pearsonr(df['x1'], df['y']) # Menggunakan 'x1' dan 'y'
print('Koefisien Korelasi Pearson (X1 dan Y):', correlation_x1_y)
print('P-value (X1 dan Y):', p_value_x1_y)

# Hitung korelasi Pearson antara X2 dan Y
correlation_x2_y, p_value_x2_y = pearsonr(df['x2'], df['y']) # Menggunakan 'x2' dan 'y'
print('Koefisien Korelasi Pearson (X2 dan Y):', correlation_x2_y)
print('P-value (X2 dan Y):', p_value_x2_y)
```

```python
#KORELASI SPEARMAN

import pandas as pd
from scipy.stats import spearmanr

# Baca file Excel dari sheet 3
data = pd.read_excel('/content/Data Korelasi Pearson [www.spssindonesia.com].xlsx', sheet_name='Sheet3')

# Buat DataFrame
df = pd.DataFrame(data)

# Mengubah nama kolom menjadi huruf kecil agar sesuai dengan data
df.columns = df.columns.str.lower()

# Hitung korelasi Spearman antara X1 dan Y
correlation_spearman_x1_y, p_value_spearman_x1_y = spearmanr(df['x1'], df['y'])
print('Koefisien Korelasi Spearman (X1 dan Y):', correlation_spearman_x1_y)
print('P-value Spearman (X1 dan Y):', p_value_spearman_x1_y)

# Hitung korelasi Spearman antara X2 dan Y
correlation_spearman_x2_y, p_value_spearman_x2_y = spearmanr(df['x2'], df['y'])
print('Koefisien Korelasi Spearman (X2 dan Y):', correlation_spearman_x2_y)
print('P-value Spearman (X2 dan Y):', p_value_spearman_x2_y)
```

```python
# UJI ASUMSI REGRESI
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

# Baca file Excel
data = pd.read_excel('/content/Data Regresi Linear Sederhana.xlsx', sheet_name='Sheet2')

# Buat DataFrame
df = pd.DataFrame(data)

# Model regresi sederhana
X = df['Stres Kerja (X)']
Y = df['Kinerja Pegawai (Y)']
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()

# Residuals
residuals = model.resid

# Uji Normalitas menggunakan Q-Q Plot
sm.qqplot(residuals, line='s')
plt.title('Q-Q Plot')
plt.show()

# Uji Linearitas dengan Scatter Plot
plt.scatter(model.fittedvalues, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted Values')
plt.show()

# Uji Multikolinearitas dengan VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

X_df = df[['Stres Kerja (X)']]
X_df = sm.add_constant(X_df)
vif_data = pd.DataFrame()
vif_data["feature"] = X_df.columns
vif_data["VIF"] = [variance_inflation_factor(X_df.values, i) for i in range(X_df.shape[1])]
print(vif_data)

# Uji Heteroskedastisitas dengan Breusch-Pagan Test
from statsmodels.stats.diagnostic import het_breuschpagan

bp_test = het_breuschpagan(residuals, X)
labels = ['Lagrange multiplier statistic', 'p-value', 'f-value', 'f p-value']
print(dict(zip(labels, bp_test)))

# Uji Autokorelasi dengan Durbin-Watson Test
dw_statistic = sm.stats.durbin_watson(residuals)
print(f'Durbin-Watson statistic: {dw_statistic}')
```

```python
#REGRESI SEDERHANA
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Baca file Excel
data = pd.read_excel('/content/Data Regresi Linear Sederhana.xlsx', sheet_name='Sheet2')

# Buat DataFrame
df = pd.DataFrame(data)

# Menentukan variabel independen dan dependen
X = df['Stres Kerja (X)']
Y = df['Kinerja Pegawai (Y)']
X = sm.add_constant(X)

# Membuat model regresi
model = sm.OLS(Y, X).fit()
print(model.summary())

# Visualisasi
plt.scatter(df['Stres Kerja (X)'], df['Kinerja Pegawai (Y)'], color='blue', label='Data Points') # Corrected column names
plt.plot(df['Stres Kerja (X)'], model.predict(X), color='red', label='Regression Line') # Corrected column names
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresi Sederhana')
plt.legend()
plt.show()

# Prediksi
value1 = 10  # Example value
value2 = 20  # Example value
new_data = pd.DataFrame({'independen': [value1, value2]})  # Ganti dengan nilai baru
new_data = sm.add_constant(new_data)
predictions = model.predict(new_data)
print(predictions)
```

```python
# REGRESI LINEAR BERGANDA
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np

# Baca file Excel
data = pd.read_excel('/content/Data Analisis Regresi Multiples-Berganda [www.spssindonesia.com].xlsx', sheet_name='Sheet1')

# Buat DataFrame
df = pd.DataFrame(data)

# Menentukan variabel independen dan dependen
X1 = df[['Motivasi (X1)', 'Minat (X2)']]
Y1 = df['Prestasi (Y)']
X1 = sm.add_constant(X1)

# Membuat model regresi berganda
model = sm.OLS(Y1, X1).fit()
print(model.summary())

# Menampilkan persamaan model regresi
intercept = model.params[0]
coef_X1 = model.params[1]
coef_X2 = model.params[2]

# Format persamaan regresi
equation = f"Prestasi (Y) = {intercept:.4f} + {coef_X1:.4f} * Motivasi (X1) + {coef_X2:.4f} * Minat (X2)"
print("Persamaan Model Regresi Linear Berganda:")
print(equation)

# Visualisasi
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Use the actual column names 'Motivasi (X1)', 'Minat (X2)', and 'Prestasi (Y)'
ax.scatter(df['Motivasi (X1)'], df['Minat (X2)'], df['Prestasi (Y)'], color='blue')

# Plotting the regression plane
# Use actual column names when generating data for the regression plane
x1_range = np.linspace(df['Motivasi (X1)'].min(), df['Motivasi (X1)'].max(), 10)
x2_range = np.linspace(df['Minat (X2)'].min(), df['Minat (X2)'].max(), 10)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)
y_grid = model.params[0] + model.params[1] * x1_grid + model.params[2] * x2_grid
ax.plot_surface(x1_grid, x2_grid, y_grid, alpha=0.5)

plt.show() # Add plt.show() to display the plot

# Prediksi
new_data = pd.DataFrame({'independen1': [value1], 'independen2': [value2]})  # Ganti dengan nilai baru
new_data = sm.add_constant(new_data)  # Tambahkan intercept
predictions = model.predict(new_data)
print(predictions)
```

```python
# REGRESI NON LINEAR LOGISTIK
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Baca file Excel
data = pd.read_excel('/content/DATA REGRESI LOGISTIK NON LINEAR.xlsx', sheet_name='Sheet1')

# Buat DataFrame
df = pd.DataFrame(data)

# Menentukan variabel independen dan dependen
X2A = df[['X1', 'X2']]
Y2A = df['Y']
X2A = sm.add_constant(X2A)

# Membuat model regresi logistik
model = sm.Logit(Y2A, X2A).fit()
print(model.summary())

# Menampilkan persamaan model regresi logistik
intercept = model.params[0]
coef_X1 = model.params[1]
coef_X2 = model.params[2]

# Format persamaan regresi logistik
equation = f"P(Y=1) = 1 / (1 + exp(-({intercept:.4f} + {coef_X1:.4f} * X1 + {coef_X2:.4f} * X2)))"
print("Persamaan Model Regresi Logistik Non-Linear:")
print(equation)

# Model regresi logistik (using X1 for prediction)
X = df['X1']  # Changed to 'X1'
Y = df['Y']
X = sm.add_constant(X)
model = sm.Logit(Y, X).fit()

# Prediksi probabilitas
df['predicted_probabilities'] = model.predict(X)

# Visualisasi
plt.scatter(df['X1'], df['Y'], color='blue', label='Data Points') # Changed to 'X1'
plt.plot(df['X1'], df['predicted_probabilities'], color='red', label='Logistic Regression Curve') # Changed to 'X1'
plt.xlabel('X1') # Changed to 'X1'
plt.ylabel('Probability of Y=1')
plt.title('Regresi Logistik')
plt.legend()
plt.show()

# Model regresi logistik (using X2 for prediction)
X = df['X2']  # Changed to 'X2'
Y = df['Y']
X = sm.add_constant(X)
model = sm.Logit(Y, X).fit()

# Prediksi probabilitas
df['predicted_probabilities'] = model.predict(X)

# Visualisasi
plt.scatter(df['X2'], df['Y'], color='blue', label='Data Points') # Changed to 'X2'
plt.plot(df['X2'], df['predicted_probabilities'], color='red', label='Logistic Regression Curve') # Changed to 'X2'
plt.xlabel('X2') # Changed to 'X2'
plt.ylabel('Probability of Y=1')
plt.title('Regresi Logistik')
plt.legend()
plt.show()

new_data = pd.DataFrame({'independen1': [value1], 'independen2': [value2]})  # Ganti dengan nilai baru
new_data = sm.add_constant(new_data)
predictions = model.predict(new_data)
print(predictions)
```

```python
#REGRESI NONLINEAR KUADRATIK
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Baca file Excel
data = pd.read_excel('/content/DATA KUADRATIK DAN KUBIK.xlsx', sheet_name='Sheet1')

# Buat DataFrame
df = pd.DataFrame(data)

# Menambahkan suku kuadratik ke DataFrame
df['X_squared'] = df['Kadar Nikotin {X}'] ** 2

# Menentukan variabel independen dan dependen
X3 = df[['Kadar Nikotin {X}', 'X_squared']]
Y3 = df['Karbon Monoksida {Y)']
X3 = sm.add_constant(X3)

# Membuat model regresi kuadratik
model = sm.OLS(Y3, X3).fit()
print(model.summary())

# Menampilkan persamaan model regresi kuadratik
intercept = model.params[0]
coef_X = model.params[1]
coef_X_squared = model.params[2]

# Format persamaan regresi kuadratik
equation = f"Karbon Monoksida (Y) = {intercept:.4f} + {coef_X:.4f} * Kadar Nikotin (X) + {coef_X_squared:.4f} * (Kadar Nikotin (X))^2"
print("Persamaan Model Regresi Non-Linear Kuadratik:")
print(equation)

# Visualisasi
x_range = np.linspace(0, 4, 100)
y_predicted = model.predict(sm.add_constant(np.column_stack((x_range, x_range**2))))

# Use the actual column names from your DataFrame:
plt.scatter(df['Kadar Nikotin {X}'], df['Karbon Monoksida {Y)'], color='blue', label='Data Points')
plt.plot(x_range, y_predicted , color='red', label='Quadratic Regression Curve')
plt.xlabel('Kadar Nikotin {X}')  # Changed to reflect the actual column name
plt.ylabel('Karbon Monoksida {Y)')  # Changed to reflect the actual column name
plt.title('Regresi Kuadratik')
plt.legend()
plt.show()

# Prediksi
new_data = pd.DataFrame({'Kadar Nikotin {X}': [value1]})  # Ganti dengan nilai baru
new_data['X_squared'] = new_data['Kadar Nikotin {X}'] ** 2
new_data = sm.add_constant(new_data)
predictions = model.predict(new_data)
print(predictions)
```

```python
#REGRESI NONLINEAR KUBIK
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Baca file Excel
data = pd.read_excel('/content/DATA KUADRATIK DAN KUBIK.xlsx', sheet_name='Sheet1')

# Buat DataFrame
df = pd.DataFrame(data)

# Menambahkan suku kubik ke DataFrame
df['X_squared'] = df['Kadar Nikotin {X}'] ** 2
df['X_cubed'] = df['Kadar Nikotin {X}'] ** 3

# Menentukan variabel independen dan dependen
X4 = df[['Kadar Nikotin {X}', 'X_squared', 'X_cubed']]
Y4 = df['Karbon Monoksida {Y)']
X4 = sm.add_constant(X4)

# Membuat model regresi kubik
model = sm.OLS(Y4, X4).fit()
print(model.summary())

# Menampilkan persamaan model regresi kubik
intercept = model.params[0]
coef_X = model.params[1]
coef_X_squared = model.params[2]
coef_X_cubed = model.params[3]

# Format persamaan regresi kubik
equation = f"Karbon Monoksida (Y) = {intercept:.4f} + {coef_X:.4f} * Kadar Nikotin (X) + {coef_X_squared:.4f} * (Kadar Nikotin (X))^2 + {coef_X_cubed:.4f} * (Kadar Nikotin (X))^3"
print("Persamaan Model Regresi Non-Linear Kubik:")
print(equation)

# Visualisasi
x_range = np.linspace(0,4 ,100)
y_predicted=model.predict(sm.add_constant(np.column_stack((x_range,x_range**2,x_range**3))))

# Use the actual column names from your DataFrame:
plt.scatter(df['Kadar Nikotin {X}'], df['Karbon Monoksida {Y)'], color='blue', label='Data Points')
plt.plot(x_range,y_predicted,color='red',label='Cubic Regression Curve')
plt.xlabel('Kadar Nikotin {X}')  # Changed to reflect the actual column name
plt.ylabel('Karbon Monoksida {Y)')  # Changed to reflect the actual column name
plt.title('Regresi Kubik')
plt.legend()
plt.show()

# Buat DataFrame baru untuk prediksi
new_data5 = pd.DataFrame({'Kadar Nikotin {X}': [2.5]})  # Ganti dengan nilai baru yang ingin diprediksi

# Tambahkan kolom kuadratik jika menggunakan regresi kuadratik
new_data5['X_squared'] = new_data5['Kadar Nikotin {X}'] ** 2

# Tambahkan kolom kubik jika menggunakan regresi kubik
new_data5['X_cubed'] = new_data5['Kadar Nikotin {X}'] ** 3

# Tambahkan intercept
new_data5 = sm.add_constant(new_data5)  # Ini menambahkan kolom konstanta (intercept)

# Lakukan prediksi
predictions = model.predict(new_data5)
print(predictions)
```
